```python
import numpy as np
import matplotlib.pyplot as plt
#%matplotlib inline

if __name__ == '__main__':
    anno_file = 'annotation.txt'
    with open(anno_file, 'r') as fp:
        lines = fp.readlines()
        
        datalist=list() #先用一個list把所有資料整理成一個list
        for line in lines:
            annotation=line.strip().split(' ') 
            data=annotation[1:]   #將每筆資料的第一筆資料拿掉(拿掉開頭)
            for number in data:
                datalist.append(number)
                
        
        npdata=np.array(datalist)
        
        npdata=npdata.reshape((int(len(npdata)/4),4))
        npdata =npdata.astype(int) #轉成int 的nparray
        #print(npdata)
        weightdata =list()
        heightdata=list()
        invalid=list()
        w1=list()
        w2=list()
        w3=list()
        w4=list()
        w5=list()
        w6=list()
        h1=list()
        h2=list()
        h3=list()
        h4=list()
        h5=list()
        h6=list()
        j=0
        for i in range(159424):
            
            if (npdata[i][0]>=0 and npdata[i][1]>=0 and npdata[i][2]>npdata[i][0] and npdata[i][3]>npdata[i][1] ):
                width =npdata[i][2]-npdata[i][0]+1
                height=npdata[i][3]-npdata[i][1]+1
                weightdata.append(width)
                heightdata.append(height)
                info=width/height

                if (width<10):
                    w1.append(width)
                elif (width>=10 and width<20):
                    w2.append(width)    
                elif (width>=20 and width<30):
                    w3.append(width)    
                elif (width>=30 and width<40):
                    w4.append(width)    
                elif (width>=40 and width<50):
                    w5.append(width)            
                elif(width>=50):
                    w6.append(width)

                if (info <0.6):
                    h1.append(info) 
                elif (info<0.7 and info>=0.6):
                    h2.append(info)
                elif(info<0.8 and info>=0.7):
                    h3.append(info)
                elif(info<0.9 and info>=0.8):
                    h4.append(info)
                elif(info<1.0 and info>=0.9):
                    h5.append(info)
                elif(info>=1.0):
                    h6.append(info) 

            else:
                j=j+1
        
        print('invalid:'+str(j))
        print('    width<10:'+str(len(w1)))
        print('10=<width<20:'+str(len(w2)))
        print('20=<width<30:'+str(len(w3)))
        print('30=<width<40:'+str(len(w4)))
        print('40=<width<50:'+str(len(w5)))
        print('    width>50:'+str(len(w6)))
        print('                   ')
        print('       width/height < 0.6:'+str(len(h1))) #1
        print('0.6 <= width/height < 0.7:'+str(len(h2))) #2
        print('0.7 <= width/height < 0.8:'+str(len(h3))) #3
        print('0.8 <= width/height < 0.9:'+str(len(h4))) #4
        print('0.9 <= width/height < 1.0:'+str(len(h5))) #5
        print('1.0 <= width/height :'+str(len(h6))) #6
        print('          ')
        
        plt.hist(weightdata, range(0,101,10), rwidth=0.85)
        plt.xlabel('Width')
        plt.ylabel('Counts')   
        plt.show()           
        
        
        
        
            

            
        
        

  # Enter your code here (25%)
  # You might use the following to access each line
  # for line in lines:
```
