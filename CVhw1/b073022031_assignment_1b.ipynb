```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from scipy import linalg
%matplotlib inline
alpha=0.02
max_iter=1000
list1=list()
list2=list()
list3=list()
def show_gradient_descent(A, b):
    x=np.array([[-2],[2]])
    print('k          (x1,x2)        f(x)' )
    for k in range(max_iter):
        a2 = np.dot(A,x)
        a3 = a2-b
        fmin = np.dot(a3.T,a3)
        print('%d' %k +'    '+'%.4f'%x[0][0]+'    '+ '%.4f' %x[1][0]+"   "+'%.8f' %fmin)
        c=np.dot(A,x)
        c2=np.dot(A.T,c)
        c3=np.dot(A.T,b)
        fx=2*(c2-c3)      
        list1.append(x[0][0])
        list2.append(x[1][0])
        list3.append(fmin[0][0])
        x_prev=x
        x=x-alpha*fx
        value=np.linalg.norm(x-x_prev)
        k=k+1
        
        if( value < 0.00000001 ):
            break
    
   
```

```python
def draw_gradient_descent(A, b):
    xx, yy = np.meshgrid(np.arange(-2,2.1,0.1), np.arange(-2,2.1,0.1))
    F1 = A[0][0]*xx + A[0][1]*yy - b[0];
    F2 = A[1][0]*xx + A[1][1]*yy - b[1];
    zz = F1*F1 + F2*F2;
    plt3d = plt.figure().gca(projection='3d')
    plt3d.plot_surface(xx, yy, zz, alpha=0.7, cmap=cm.coolwarm)
    
    show_gradient_descent( A , b )
    ax = plt.gca(projection='3d')

    for k in range(37):
        ax.scatter(list1[k],list2[k],list3[k],c='r')
        
    x = np.arange(-2, 2 , 0.05 )
    y=0.5*x-0.75
    plt3d.plot(x, y,color="blue")
    plt.show()      


    plt.show()
```

```python
if __name__ == '__main__':
    b = np.array([[1], [2]])
    A = np.array([[1, -2], [1, -2]])
  
    draw_gradient_descent(A, b)
```
